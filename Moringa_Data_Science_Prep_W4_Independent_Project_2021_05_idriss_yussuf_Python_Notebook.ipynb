```python
#impoting pandas library
import pandas as pd
# importing numpy library
import numpy as np
```

```python
#loading the data set
df= pd.read_csv('http://bit.ly/autolib_dataset')
df.head(3)
```

```python
#checking the info of the data
df.info()
```

```python
#checking the shape of the data
df.shape
```

```python
#cecking the missing values from the data
df.isnull().sum()
```

```python
#dropping the columns that are not useful for the analysis
df.drop(['Geo point','Displayed comment'],axis =1,inplace=True)
df.head(2)
```

```python
#checking the shape again to confirm if the two columns are dropped
df.shape
```

```python
#creating date_time column in our dataset by combining year,month,day,hour
df['date_time'] = df['year'].map(str) +':'+ df['month'].map(str) +':'+ df['day'].map(str) +':'+ df['hour'].map(str) +':'+ df['minute'].map(str)
df.head(3)
```

```python
#dropping the columns year,month day minutes
df.drop(['year','month','day','minute'], axis = 1,inplace=True)
df.head(3)
```

```python
#checking the shape to confirm if the columns have been dropped
df.shape
```

```python
#import datetime and change the date_time column format
import datetime as datetime
df['date_time'] = pd.to_datetime(df['date_time'] , format='%Y:%m:%d:%H:%M')
df.head(2)
```

```python
#selecting the city as Paris
df2= df.loc[df['City'] == 'Paris']
df2.head(3)
```

```python
#checking the shape of the paris data
df2.shape
```

```python
#grouping together
df3 = df2.groupby(['ID','date_time','hour']).sum()
df3
```

```python
#getting the difference of the bluecar counter to check if the cars were rented or returned
df3['difference'] = df3['Bluecar counter'].diff()
df3.head(3)
```

```python
#selecting where the difference is greater than 0,to indicate a car was picked
df4 = df3[df3['difference'] > 0]
df4
```

```python
#grouping the difference in the bluecar counter and the Hour
df4.groupby(['hour'])['difference'].sum().sort_values(ascending = False)
```
